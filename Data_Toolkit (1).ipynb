```python
#np.array: This function creates a new NumPy array from an input data structure (like a list or tuple). It always creates a new copy of the data, even if the input is already a NumPy array.

#np.asarray: This function also creates a new NumPy array from an input data structure, but it may return a view of the input data instead of creating a new copy. This means that changes made to the input data may also affect the array created with np.asarray.

#np.asanyarray: This function also creates a new NumPy array, but it will try to avoid making a copy of the input data whenever possible. If the input is already a NumPy array, np.asanyarray will return a view of that array instead of creating a new copy. If the input is not a NumPy array, it will create a new array.
```

```python
#A deep copy and a shallow copy are two different ways of creating copies of an object or data structure.

#A shallow copy creates a new object or data structure, but only copies the top-level structure of the original object. In other words, it creates a new object with references to the original objects, rather than creating copies of those objects themselves. This means that changes made to the original object may also affect the shallow copy.

#On the other hand, a deep copy creates a new object or data structure and then recursively copies all of the nested objects or elements within the original object. This means that changes made to the original object will not affect the deep copy, as they are completely separate entities.

#In summary, a shallow copy only creates a copy of the top-level structure of the object, while a deep copy creates copies of all nested objects or elements as well.
```

```python
#Create a 5x6 array with random integers between 1 and 10
array = np.random.randint(1, 11, (5, 6))

print("Original Array:")
print(array)

# Extract all even integers from the array
even_numbers = array[array % 2 == 0]

print("\nEven Integers:")
print(even_numbers)

# Extract all odd integers from the array
odd_numbers = array[array % 2 != 0]

print("\nOdd Integers:")
print(odd_numbers)
```

```python
import numpy as np

# Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10
array = np.random.randint(1, 11, (3, 3, 3))
print("Original Array:")
print(array)

# Find the indices of the maximum values along each depth level (third axis)
max_indices = np.argmax(array, axis=2)
print("\nIndices of Maximum Values along Depth Level:")
print(max_indices)

# Perform element-wise multiplication between both arrays
result = array * max_indices[:,:,np.newaxis]
print("\nElement-wise Multiplication Result:")
print(result)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Create two NumPy arrays containing 100 random float values between 0 and 1
x = np.random.rand(100)
y = np.random.rand(100)

#Create the scatter plot
plt.scatter(x, y, color='red', marker='o', label='Random Values')

#Add horizontal line at y = 0.5
plt.axhline(y=0.5, color='black', linestyle='--', label='y = 0.5')

#Add vertical line at x = 0.5
plt.axvline(x=0.5, color='black', linestyle=':', label='x = 0.5')

#Label x-axis and y-axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

#Set title of the plot
plt.title('Advanced Scatter Plot of Random Values')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a time-series dataset
np.random.seed(0)
dates = pd.date_range('20220101', periods=100)
temperature = np.random.randint(60, 100, 100)
humidity = np.random.randint(40, 80, 100)

df = pd.DataFrame({'Date': dates,
                   'Temperature': temperature,
                   'Humidity': humidity})

# Plotting Temperature and Humidity on the same plot with different y-axes
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity', color=color)
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(size=1000)

count, bins, ignored = plt.hist(data, 30, density=True, alpha=0.5, color='b', edgecolor='black')
plt.plot(bins, 1/(np.sqrt(2 * np.pi)) * np.exp( - (bins**2) / 2), linewidth=2, color='r')

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays of size 100
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
quadrants = []
for i in range(100):
    if x[i] > 0 and y[i] > 0:
        quadrants.append("Quadrant 1")
    elif x[i] < 0 and y[i] > 0:
        quadrants.append("Quadrant 2")
    elif x[i] < 0 and y[i] < 0:
        quadrants.append("Quadrant 3")
    elif x[i] > 0 and y[i] < 0:
        quadrants.append("Quadrant 4")

# Create a dataframe for the data
data = {
    'x': x,
    'y': y,
    'Quadrant': quadrants
}
df = pd.DataFrame(data)

# Create a scatter plot with color points based on their quadrant
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette=['blue', 'red', 'green', 'orange'])
plt.legend(title='Quadrant')
plt.xlabel('X-axis')
plt
```

```python
from bokeh.plotting import figure, show
import numpy as np


# Generating data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Creating a figure
p=figure(title='Sine Wave Function', x_axis_label='x', y_axis_label='sin(x)')

# Adding grid lines
p.grid.grid_line_color = 'gray'
p.grid.grid_line_dash = [6, 4]

# Adding line plot
p.line(x, y, line_width=2)

# Display the plot
show(p)
```

```python
from bokeh.plotting import figure,show
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6
import random

# generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
data = {cat: random.randint(1, 10) for cat in categories}

# create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=list(data.values())))

# create a figure
r = figure(x_range=categories,height=350,title='Random Categorical Bar Chart',
           toolbar_location=None, tools="")

# add bar chart
r.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color=factor_cmap('categories', palette=Spectral6, factors=categories))

# add hover tooltips
r.add_tools(HoverTool(tooltips=[('Category', '@categories'), ('Value', '@values')]))

# set labels
r.xaxis.axis_label = 'Categories'
r.yaxis.axis_label = 'Values'

show(r)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import HoverTool
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, size=len(categories))

output_notebook()
p = figure(x_range=categories, height=350, title='Random Categorical Bar Chart',
           toolbar_location=None, tools='')

colors = ['#FF5733', '#33FF57', '#337AFF', '#FF33E9', '#FFD633']
p.vbar(x=categories, top=values, width=0.9, color=colors)

hover = HoverTool()
hover.tooltips = [('Value', '@top')]
p.add_tools(hover)

p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

show(p)
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Create a DataFrame with randomly generated data
data = {'labels': ['A', 'B', 'C', 'D', 'E'],
        'values': np.random.randint(1, 100, 5)}

df = pd.DataFrame(data)

# Calculate percentage
df['percentage'] = df['values'] / df['values'].sum() * 100

# Create an interactive pie chart
fig = px.pie(df, values='values', names='labels', labels={'labels':'Label', 'values':'Value', 'percentage':'Percentage'})

fig.update_traces(textposition='inside', textinfo='percent+label')

fig.update_layout(title='Interactive Pie Chart')

fig.show()
```
